```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Reading in CSVs

Merge_2019 = pd.read_csv("MERGE_2019_df.csv")
Merge_2018 = pd.read_csv("YR2018Data.csv")
Merge_2017 = pd.read_csv("YR2017Data.csv")
Merge_2016_start = pd.read_csv("rdu_2016_data_AD.csv")
Merge_2015_start = pd.read_csv("rdu_2015_data_AD.csv")

L_AIRPORT_ID = pd.read_csv("L_AIRPORT_ID.csv")
L_UNIQUE_CARRIERS = pd.read_csv("L_UNIQUE_CARRIERS.csv")
RDUCurrentDestinations_start = pd.read_csv("RDUCurrentDestinations.csv")
RDUCurrentFlights_start = pd.read_csv("RDUCurrentFlights.csv")
RDU_2015_19_Delays_Causes_ML = pd.read_csv("RDU_2015_19_Delays_Causes_ML.csv")
```

```python
Merge_2016_start['ORIGIN_AIRPORT_ID']='14492'
Merge_2015_start['ORIGIN_AIRPORT_ID']='14492'
Merge_2015=Merge_2015_start.drop('Unnamed: 0',axis=1)
Merge_2016=Merge_2016_start.drop('Unnamed: 0',axis=1)
```

```python
# Created merged Dataframe. Use .append() to add each quarter

FINAL_MERGE_df = Merge_2019.append(Merge_2018).append(Merge_2017).append(Merge_2016).append(Merge_2015)

# Display data for the merged data

FINAL_MERGE_df
```

```python
RDUCurrentDestinations=RDUCurrentDestinations_start.drop('Unnamed: 0',axis=1)
RDUCurrentFlights=RDUCurrentFlights_start.drop('Unnamed: 0',axis=1)
```

```python
connection_string = f'postgresql://postgres:Syracuse10!@localhost:5432/where-can-we-go'
engine = create_engine(connection_string)
```

```python
engine.table_names()
```

```python
FINAL_MERGE_df.to_sql(name='FINAL_MERGE', con=engine, if_exists='replace')
pd.read_sql_query('select * from "FINAL_MERGE"', con=engine).head()
```

```python
L_AIRPORT_ID.to_sql(name='L_AIRPORT_ID', con=engine, if_exists='replace')
pd.read_sql_query('select * from "L_AIRPORT_ID"', con=engine).head()
```

```python
L_UNIQUE_CARRIERS.to_sql(name='L_UNIQUE_CARRIERS', con=engine, if_exists='replace')
pd.read_sql_query('select * from "L_UNIQUE_CARRIERS"', con=engine).head()
```

```python
RDUCurrentDestinations.to_sql(name='RDUCurrentDestinations', con=engine, if_exists='replace')
pd.read_sql_query('select * from "RDUCurrentDestinations"', con=engine).head()
```

```python
RDUCurrentFlights.to_sql(name='RDUCurrentFlights', con=engine, if_exists='replace')
pd.read_sql_query('select * from "RDUCurrentFlights"', con=engine).head()
```

```python
RDU_2015_19_Delays_Causes_ML.to_sql(name='RDU_2015_19_Delays_Causes_ML', con=engine, if_exists='replace')
pd.read_sql_query('select * from "RDU_2015_19_Delays_Causes_ML"', con=engine).head()
```
