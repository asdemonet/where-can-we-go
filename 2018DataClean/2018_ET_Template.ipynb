```python
# Import Pandas

import pandas as pd
```

```python
# Read in CSVs to Dataframes

# !! Important. For each quarter change the file name to read in the correct quarter #

Coup2018Q1_df = pd.read_csv('RawData_CoupTix_2018/Coup2018Q4.csv')
Tix2018Q1_df = pd.read_csv('RawData_CoupTix_2018/Tix2018Q4.csv')
```

```python
# Preview Coupon 2018 Dataframe

Coup2018Q1_df.head()
```

```python
# Check Columns

Coup2018Q1_df.columns
```

```python
# Filter by Origin = RDU

Coup2018Q1_df = Coup2018Q1_df.loc[(Coup2018Q1_df["ORIGIN"] == "RDU"),:]

# Check Dataframe and Drop Unnamed Column

Coup2018Q1_df = Coup2018Q1_df.drop('Unnamed: 15', axis=1)
```

```python
# Preview Tix 2018 Dataframe

Tix2018Q1_df.head()
```

```python
# Removed Unnamed Column

Tix2018Q1_df = Tix2018Q1_df.drop('Unnamed: 4', axis=1)

Tix2018Q1_df.head()
```

```python
# Merge DataFrames on Itinerary ID and drop null values

Combined2018Q1_df = Coup2018Q1_df.merge(Tix2018Q1_df, on = 'ITIN_ID')
Combined2018Q1_df.dropna()
Combined2018Q1_df.head()
#Combined2018Q1_df.dtypes()
```

```python
# Filter for Non-Stop roundtrip flights only (Coupons = 2 & RoundTrip = 1)

Combined2018Q1_df_final = Combined2018Q1_df.loc[(Combined2018Q1_df["COUPONS"] == 2) & (Combined2018Q1_df["ROUNDTRIP"] == 1.0),:]
Combined2018Q1_df_final
```

```python
# !!!! IMPORTANT !!!!! Make Sure to change CSV file name to save separate

Combined2018Q1_df_final.to_csv('CleanQuarterCSVs/Combined2018Q4',index=False)
```
